```python
from keybert import KeyBERT
from sklearn.feature_extraction.text import CountVectorizer

# KeyBERT 모델 초기화 (klue 사용)
kw_model = KeyBERT(model='klue/roberta-base')

def extract_keywords_from_reviews(reviews, top_n=5):
    # 주어진 리뷰들의 문장 리스트에서 각 문장별로 키워드를 추출하여 출력
    
    for review in reviews:
        print(f"Review: {review}\n")
        for sentence in review:
            # 키워드 추출
            keywords = kw_model.extract_keywords(sentence, 
                                                 keyphrase_ngram_range=(1, 1),  # 단어 n-gram 범위
                                                 stop_words=None,  # 불용어
                                                 use_maxsum=False, use_mmr=True, diversity=0.7,  # 다양성
                                                 top_n=top_n)  # 상위 n개 키워드
            print(f"Sentence: {sentence}\nKeywords: {keywords}\n")
        print("-" * 80)  # 리뷰 사이 구분선

# 처음 10개 식당의 문장 리스트 추출
review_sentences = data['review_sentence_split'].iloc[:10].tolist()

# 키워드 추출 실행
extract_keywords_from_reviews(review_sentences)
```

```python
from keybert import KeyBERT
from sklearn.feature_extraction.text import CountVectorizer

# KeyBERT 모델 초기화 (klue 사용)
kw_model = KeyBERT(model='klue/roberta-base')

def extract_keywords_from_reviews(reviews, top_n=5):
    # 주어진 리뷰들의 문장 리스트에서 각 문장별로 키워드를 추출하여 출력
    
    for review in reviews:
        print(f"Review: {review}\n")
        
        keywords = kw_model.extract_keywords(review,
                                             keyphrase_ngram_range=(1, 1),  # 단어 n-gram 범위
                                             stop_words=None,  # 불용어
                                             use_maxsum=False, use_mmr=True, diversity=0.7,  # 다양성
                                             top_n=top_n)  # 상위 n개 키워드)
        print(f"Sentence: {review}\nKeywords: {keywords}\n")
    print("-" * 80)  # 리뷰 사이 구분선


# 처음 10개 식당의 문장 리스트 추출
review_sentences = data['review_sentence_split'].iloc[:10].tolist()

# 키워드 추출 실행
extract_keywords_from_reviews(review_sentences)
```
