```python
# import library
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline

file_path = '/Users/gimtaeho/Desktop/Sentiment-Classification/koelectra/menu_sentimental_ver1.csv'
data = pd.read_csv(file_path, nrows=100)
```

```python
# load model
tokenizer = AutoTokenizer.from_pretrained("jaehyeong/koelectra-base-v3-generalized-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("jaehyeong/koelectra-base-v3-generalized-sentiment-analysis")
sentiment_classifier = TextClassificationPipeline(tokenizer=tokenizer, model=model)
```

```python
# 감성분석 함수
def get_sentiment_score(sentence):
    result = sentiment_classifier(sentence)[0]
    return result['score'] if result['label'] == '1' else -result['score']

# 메뉴 센텐스 열 갖고 센티 스코어 만들기
data['senti_score'] = data['menu_sentence'].apply(get_sentiment_score)
```

```python
# 각 문장에 대한 모델의 예측 결과를 확인
predictions = sentiment_classifier(data['menu_sentence'].tolist())
for pred in predictions:
    print(pred)
```
